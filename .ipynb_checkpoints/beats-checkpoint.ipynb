```python
import os
import pandas as pd
import google.generativeai as genai
from textblob import TextBlob # For sentiment analysis

# --- Configuration for Gemini API ---
# IMPORTANT: It's highly recommended to use environment variables for API keys in production
# (e.g., os.getenv("GEMINI_API_KEY")) instead of hardcoding.
API_KEY = "user_change"
genai.configure(api_key=API_KEY)

# Set up model configuration for general use
GLOBAL_GENERATION_CONFIG = {
  "temperature": 1,             # Controls creativity (1 = more creative, 0 = more focused)
  "top_p": 0.95,                # Nucleus sampling
  "top_k": 64,                  # Top-K sampling
  "max_output_tokens": 8192,    # Maximum length of responses
  "response_mime_type": "text/plain", # Format of the response
}

# Initialize the model with global configuration
global_model = genai.GenerativeModel(
  model_name="gemini-1.5-flash",
  generation_config=GLOBAL_GENERATION_CONFIG
)

# --- Step 1: Understanding the Gemini Setup (Quick Experimentation) ---
# Initialize a chat session with an initial context
initial_chat_session = global_model.start_chat(
    history=[{
        "role": "user",
        "parts": ["You are a market research analyst specializing in consumer electronics."]
    }]
)

print("--- Quick Experimentation ---")
response1 = initial_chat_session.send_message("What features matter most to headphone buyers?")
print("Response 1:", response1.text)

response2 = initial_chat_session.send_message("How would you analyze customer feedback about sound quality?")
print("Response 2:", response2.text)
print("-" * 30)


# --- Step 2 & 3.0: Core Review Analysis Function ---
def analyze_reviews(reviews_list, prompt_message_str):
    """
    Analyzes a list of reviews using Gemini AI based on a given prompt.
    Creates a new chat session for each call to ensure isolation for different analysis tasks.
    """
    model_for_analysis = genai.GenerativeModel(
        model_name="gemini-1.5-flash",
        generation_config=GLOBAL_GENERATION_CONFIG,
    )
    chat_session_for_analysis = model_for_analysis.start_chat(history=[])
    prompt = prompt_message_str + "\n\n" + "\n\n".join(reviews_list)
    response = chat_session_for_analysis.send_message(prompt)
    return response.text

# --- Step 2: Working with Review Data ---
print("\n--- Step 2: Working with Review Data ---")
df = None # Initialize df to None
try:
    df = pd.read_csv('cleaned_dataset.csv')
    print("Dataset loaded successfully.")
    print("Columns:", df.columns.tolist())

    # Verify required columns AFTER successful loading
    REQUIRED_CONTENT_COL = 'content'
    REQUIRED_PRODUCT_ID_COL = 'product_id'

    if REQUIRED_CONTENT_COL not in df.columns:
        print(f"Error: Required column '{REQUIRED_CONTENT_COL}' not found for reviews. Please check your dataset.")
        exit()
    if REQUIRED_PRODUCT_ID_COL not in df.columns:
        print(f"Error: Required column '{REQUIRED_PRODUCT_ID_COL}' not found for product IDs. Please check your dataset.")
        exit()

except FileNotFoundError:
    print("Error: 'cleaned_dataset.csv' not found. Make sure the file is in the correct directory.")
    print("Please upload 'cleaned_dataset.csv' to the Colab environment or provide the correct path.")
    exit()
except Exception as e:
    print(f"An error occurred while reading 'cleaned_dataset.csv': {e}")
    exit()

# The rest of your code that uses df should follow here
reviews_for_analysis = df[REQUIRED_CONTENT_COL].dropna().head(30).tolist()
print(f"Selected {len(reviews_for_analysis)} reviews for initial analysis.")

prompt_basic_analysis = "Analyze these reviews and summarize key themes and customer opinions:"
insights_basic = analyze_reviews(reviews_for_analysis, prompt_basic_analysis)
print("\n--- Basic Analysis Insights ---")
print(insights_basic)
print("-" * 30)

# --- Step 3.1: Organizing Product Information ---
print("\n--- Step 3.1: Organizing Product Information ---")

# Provided product mapping from user's input (product_id: short_name)
raw_product_list = """
B09PYVXXW5: JBL Quantum TWS Earbuds
B09H27SXHS: Sony XM4 Noise Cancelling Earbuds
B09FM6PDHP: Beats Fit Pro Earbuds
B09HN594TL: JBL Tune 230NC Headphones
B096SV8SJG: Beats Studio Buds
B0BTZKP1TP: Sony WF-C700N Earbuds
B08ZR5JB9G: Sennheiser CX True Earbuds
B0C2F5KD26: Soundcore Space A40 Earbuds
B0BYPFNW6T: JBL Vibe Beam Headphones
B093SLWMS7: Soundcore Life P3 Earbuds
B0CXL4FQBK: JBL Tune Buds Headphones
B099TJGJ91: Bose SoundLink Flex Speaker
B0D4SX9RC6: JBL Wave Buds Headphones
B08X4YMTPM: JBL Charge 5 Speaker
B0B44F1GGK: Soundcore Sport X10 Earbuds
B0BG93TZ5N: Apple AirPods Pro 2nd Gen
B09XXW54QG: JBL Tune 710BT Headphones
B07DD3WBYW: Ultimate Ears Boom 3
B07YBN9XXG: Bose SoundLink Mini II
B0CY6S748H: JBL Tune Beam Headphones
B01IOD7KB6: House of Marley Get Together
B0CVFM97GD: JBL Go 4 Speaker
B08NLCW9WY: JBL Xtreme 3 Speaker
B08YRT9T38: B&O Beosound Explore
B08VL5S148: Bose SoundLink Revolve+ II
B097XX34SL: Monster S320 Speaker
B07PXGQC1Q: Apple AirPods 2nd Gen
B0BZ9WMLNQ: JBL Tune 520BT Headphones
B0D3JB14QS: JBL Vibe Buds Headphones
B0BQPNMXQV: Sony LinkBuds S Earbuds
B0C1QWWZR4: JBL Clip 4 Speaker
B09CKF166Y: Samsung Galaxy Buds2
B0CF7GYNW2: JBL Vibe Flex Headphones
B09CFP6J6D: Sony WF-C500 Earbuds
B0BZTCXG6T: Anker Life Q20 Headphones
B0863H1JKB: Panasonic RZ-S300W Earbuds
B09JL41N9C: JBL Wave 100TWS Earbuds
B09V9P5Q6W: JBL Live Free 2 Earbuds
B0BZK2Z2TC: JBL Quantum 100 Headset
B0C345M3T7: JBL Wave Buds White
B0CD2FSRDD: JBL Tune 720BT Headphones
B0CPFV77W4: Anker Soundcore P20i Earbuds
B0B1NGPY94: Sony WH-CH520 Headphones
B0BYSQDWRT: JBL Wave Beam White
B0B2SH4CN6: JBL Go 3 Speaker
B09D1HMBQ3: JBL Tune 710BT White
B0D4STD5ZC: JBL Wave Buds Blue
B099TLMRB6: Bose SoundLink Flex Blue
B07GWRCZQP: JBL Charge 4 Speaker
B09GK5JMHK: Google Pixel Buds A-Series
B088KRKFJ3: Marshall Stockwell II Speaker
B08Z1RP9K8: Sonos Roam Smart Speaker
B0B445JCZ3: JBL Quantum 350 Headset
B0B43Y8GHZ: JBL Quantum 610 Headset
"""

product_mapping = {}
brand_names_to_extract = ["JBL", "Sony", "Beats", "Sennheiser", "Soundcore", "Bose",
                          "Apple", "Ultimate Ears", "House of Marley", "B&O", "Monster",
                          "Samsung", "Anker", "Panasonic", "Google", "Marshall", "Sonos"]

for line in raw_product_list.strip().split('\n'):
    if line.strip():
        parts = line.strip().split(': ', 1)
        if len(parts) == 2:
            product_id = parts[0]
            short_name = parts[1]

            # Infer brand from the short_name
            brand = "Unknown"
            for b_name in brand_names_to_extract:
                if short_name.startswith(b_name):
                    brand = b_name
                    break

            product_mapping[product_id] = {"short_name": short_name, "brand": brand}

print("Product mapping created with brands:")
for pid, info in list(product_mapping.items())[:5]: # Print first 5 for sanity check
    print(f"- {pid}: {info['short_name']} (Brand: {info['brand']})")
print("...")

# Function to apply the mapping to your dataset
def clean_product_info(df_input, mapping):
    df_input['short_name'] = df_input[REQUIRED_PRODUCT_ID_COL].map({k: v['short_name'] for k, v in mapping.items()})
    df_input['brand'] = df_input[REQUIRED_PRODUCT_ID_COL].map({k: v['brand'] for k, v in mapping.items()})

    # Handle any products not in mapping
    df_input['short_name'] = df_input['short_name'].fillna(df_input[REQUIRED_PRODUCT_ID_COL])
    df_input['brand'] = df_input['brand'].fillna('Unknown')

    return df_input

# Clean your dataset by applying the product mapping
# Make sure df is not None before calling clean_product_info
if df is not None:
    df = clean_product_info(df.copy(), product_mapping)
    print("\nSample of cleaned data with short names and brands:")
    print(df[[REQUIRED_CONTENT_COL, 'short_name', 'brand']].head())
    print("-" * 30)
else:
    print("DataFrame was not loaded successfully. Cannot proceed with cleaning product info.")


# --- Step 3.2: Adding Sentiment Information ---
print("\n--- Step 3.2: Adding Sentiment Information ---")

def add_sentiment_info(df_input):
    df_input['sentiment_score'] = df_input[REQUIRED_CONTENT_COL].apply(lambda x: TextBlob(str(x)).sentiment.polarity)
    df_input['sentiment_category'] = df_input['sentiment_score'].apply(lambda x:
        'Positive' if x > 0.1
        else 'Negative' if x < -0.1
        else 'Neutral')
    return df_input

# Make sure df is not None before adding sentiment info
if df is not None:
    df = add_sentiment_info(df.copy())
    print("\nSentiment Distribution:")
    print(df['sentiment_category'].value_counts())
    print("-" * 30)
else:
    print("DataFrame was not loaded successfully. Cannot proceed with adding sentiment info.")


# --- Step 3.3: Basic Product Analysis ---
print("\n--- Step 3.3: Basic Product Analysis ---")

def analyze_basic_product(reviews_list, product_name_str):
    prompt_message = f"""Analyze these reviews for {product_name_str}:
    1. What are the top 3 most mentioned features?
    2. What do customers love most about this product?
    3. What are the most common complaints?
    Please provide specific examples from the reviews."""
    # Ensure analyze_reviews does not require an API key parameter here as per its definition
    return analyze_reviews(reviews_list, prompt_message)

# Example: Analyze a random Beats product from your list
if df is not None:
    beats_products = df[df['brand'] == 'Beats']['short_name'].unique().tolist()
    if beats_products:
        product_to_analyze_basic = beats_products[0] # Pick the first Beats product
        product_reviews_basic = df[df['short_name'] == product_to_analyze_basic][REQUIRED_CONTENT_COL].dropna().tolist()[:30]

        if product_reviews_basic:
            basic_product_insights = analyze_basic_product(product_reviews_basic, product_to_analyze_basic)
            print(f"\n--- Basic Insights for {product_to_analyze_basic} ---")
            print(basic_product_insights)
        else:
            print(f"\nNo reviews found for '{product_to_analyze_basic}'.")
    else:
        print("\nNo Beats products found in your dataset for basic analysis example.")
else:
    print("DataFrame was not loaded successfully. Cannot perform basic product analysis.")
print("-" * 30)


# --- Step 3.4: Sentiment-Based Analysis ---
print("\n--- Step 3.4: Sentiment-Based Analysis ---")

def analyze_sentiment_patterns(reviews_list, sentiment_category_str, product_name_str):
    prompt_message = f"""Analyze these {sentiment_category_str.lower()} reviews for {product_name_str}:
    1. What specific aspects create {sentiment_category_str.lower()} reactions?
    2. Identify common phrases or descriptions used
    3. Note any comparisons with competitor products
    4. Find unexpected or unique feedback
    Support your analysis with specific examples."""
    # Ensure analyze_reviews does not require an API key parameter here as per its definition
    return analyze_reviews(reviews_list, prompt_message)

# Example: Analyze positive reviews for a selected product
if df is not None:
    if beats_products: # Reuse the product picked for basic analysis
        positive_reviews_product = df[
            (df['short_name'] == product_to_analyze_basic) &
            (df['sentiment_category'] == 'Positive')
        ][REQUIRED_CONTENT_COL].dropna().tolist()[:30]

        if positive_reviews_product:
            positive_insights = analyze_sentiment_patterns(
                positive_reviews_product,
                'Positive',
                product_to_analyze_basic
            )
            print(f"\n--- Positive Insights for {product_to_analyze_basic} ---")
            print(positive_insights)
        else:
            print(f"\nNo positive reviews found for '{product_to_analyze_basic}'.")
    else:
        print("\nCannot perform sentiment analysis without a selected product.")
else:
     print("DataFrame was not loaded successfully. Cannot perform sentiment analysis.")
print("-" * 30)


# --- Step 3.5: Comparative Analysis ---
print("\n--- Step 3.5: Comparative Analysis ---")

def analyze_comparative_reviews(reviews_dict_input):
    products_list_str = ", ".join(reviews_dict_input.keys())
    prompt_message = f"""Compare customer feedback across these products: {products_list_str}

    For each product, analyze:
    1. Standout features and unique selling points
    2. Common complaints or limitations
    3. Price-to-value perception
    4. Target user preferences

    Then compare:
    1. Key differences in customer satisfaction
    2. Feature preferences across products
    3. Quality and reliability mentions

    Format as a structured comparison with specific examples."""

    combined_reviews = []
    for product, reviews in reviews_dict_input.items():
        # Ensure we only take up to 20 reviews per product for comparison
        labeled_reviews = [f"[{product}] {review}" for review in reviews[:20]]
        combined_reviews.extend(labeled_reviews)
    # Ensure analyze_reviews does not require an API key parameter here as per its definition
    return analyze_reviews(combined_reviews, prompt_message)

# Example: Compare a Beats product with a Sony and a JBL product from your list
# Ensure these products have reviews in your dataset after mapping
if df is not None:
    products_to_compare = []
    example_brands = {
        'Beats': None,
        'Sony': None,
        'JBL': None
    }

    for brand_name in example_brands.keys():
        products_of_brand = df[df['brand'] == brand_name]['short_name'].unique().tolist()
        if products_of_brand:
            example_brands[brand_name] = products_of_brand[0] # Pick the first product of that brand
            products_to_compare.append(products_of_brand[0])
        else:
            print(f"Warning: No products found for brand '{brand_name}' in dataset for comparison example.")

    reviews_dict_for_comparison = {}
    for product_comp in products_to_compare:
        reviews_comp = df[df['short_name'] == product_comp][REQUIRED_CONTENT_COL].dropna().tolist()
        if reviews_comp:
            # Limit to 20 reviews per product for the comparative analysis example
            reviews_dict_for_comparison[product_comp] = reviews_comp[:20]
        else:
            print(f"Warning: No reviews found for '{product_comp}' for comparative analysis.")

    if reviews_dict_for_comparison and len(reviews_dict_for_comparison) >= 2:
        comparison_insights = analyze_comparative_reviews(reviews_dict_for_comparison)
        print("\n--- Comparative Analysis Insights ---")
        print(comparison_insights)
    else:
        print("\nNot enough products with reviews found for comparative analysis (needs at least 2).")
else:
    print("DataFrame was not loaded successfully. Cannot perform comparative analysis.")
print("-" * 30)


# --- Step 3.6: Feature-Specific Analysis ---
print("\n--- Step 3.6: Feature-Specific Analysis ---")

def analyze_feature(reviews_list, feature_name_str):
    prompt_message = f"""Analyze mentions of {feature_name_str} in these reviews:
    1. How do customers describe the {feature_name_str}?
    2. What specific aspects are praised or criticized?
    3. How does {feature_name_str} compare across products?
    4. What improvements do customers suggest?
    Include specific quotes and examples."""
    # Ensure analyze_reviews does not require an API key parameter here as per its definition
    return analyze_reviews(reviews_list, prompt_message)

# Example: Analyze battery life mentions
# Adjust keywords for the specific feature you want to analyze
if df is not None:
    feature_keywords = 'battery|charge|power|life'
    feature_reviews = df[
        df[REQUIRED_CONTENT_COL].fillna('').str.contains(feature_keywords, case=False, na=False)
    ][REQUIRED_CONTENT_COL].dropna().tolist()[:40]

    if feature_reviews:
        battery_insights = analyze_feature(feature_reviews, "battery life")
        print("\n--- Feature-Specific Analysis (Battery Life) ---")
        print(battery_insights)
    else:
        print(f"\nNo reviews found mentioning keywords '{feature_keywords}' for feature analysis.")
else:
    print("DataFrame was not loaded successfully. Cannot perform feature-specific analysis.")
print("-" * 30)

print("\n--- Analysis Complete ---")
print("You can now review the printed insights and begin synthesizing them into your Competitive Analysis Report.")
print("Feel free to adjust the 'product_to_analyze_basic', 'products_to_compare', and 'feature_keywords' variables to target specific products or features of interest.")
```
