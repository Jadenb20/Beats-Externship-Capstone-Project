```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from textblob import TextBlob
from wordcloud import WordCloud
from google.colab import drive
from google import genai
from google.colab import userdata
```

```python
drive.mount('/content/drive')
file_path = "/content/drive/MyDrive/wireless_speakers_cleaned.csv"
```

```python
df = pd.read_csv(file_path)
```

```python
features= ["sound_quality","battery_life","design","bluetooth","durability","price"]
```

```python
missing_features = [c for c in features if c not in df.columns]
if missing_features:
    raise ValueError(f"Missing feature columns in CSV: {missing_features}")
```

```python
if "improvement_text" not in df.columns:
    raise ValueError("Missing required column: improvement_text")
```

```python
def rank_features(df, features):
    """
    Rank speaker features by average importance score.
    """
    return df[features].mean().sort_values(ascending=False)
```

```python
print(rank_features(df, features))
```

```python
def add_overall_score(df,features, new_col="overall_score"):

    """
    Calculate overall score per person for sentiment analysis axis=1 for rows
    """
    df[new_col] = df[features].mean(axis=1)
    return df
```

```python
add_overall_score(df,features, new_col="overall_score")
```

```python
df = add_overall_score(df,features)
```

```python
def segmentation(df,high_cutoff=4,low_cutoff=3,score_column="overall_score"):
    """
    Seperate the higher end of rating and the lower end
    """
    high_df = df[df[score_column]>= high_cutoff]
    low_df = df[df[score_column]<= low_cutoff]
    return high_df, low_df
```

```python
# Clean and standardize improvement feedback text, then analyze sentiment
# Polarity measures how positive or negative the feedback is (-1 to 1)
# Subjectivity measures how opinion-based vs factual the feedback is (0 to 1)
df["improvement_text_clean"] = df["improvement_text"].fillna("").astype(str)
df["improvement_polarity"] = df["improvement_text_clean"].apply(lambda t: TextBlob(t).sentiment.polarity)
df["improvement_subjectivity"] = df["improvement_text_clean"].apply( lambda t: TextBlob(t).sentiment.subjectivity)
```

```python
# Categorize sentiment into buckets using polarity thresholds
# This simplifies continuous sentiment scores into interpretable labels
# (Positive, Neutral, Negative) for downstream analysis and comparison
df["sentiment_bucket"] = np.where(
    df["improvement_polarity"] > 0.1, "Positive",
    np.where(df["improvement_polarity"] < -0.1, "Negative", "Neutral")
)
```

```python
highly_satisfied, unsatisfied = segmentation(df)
```

```python
highly_satisfied.head()
```

```python
unsatisfied.head()
```

```python
df["sentiment_bucket"].value_counts()
```

```python
improvement_text = df["improvement_text"].dropna()
happy_text = highly_satisfied["improvement_text"].dropna()
unhappy_text = unsatisfied["improvement_text"].dropna()
happy_text_list = happy_text.tolist()
unhappy_text_list = unhappy_text.tolist()
```

```python
def build_text_blob(text_series):
    return " ".join(text_series.tolist())
```

```python
happy_blob = build_text_blob(happy_text)
unhappy_blob = build_text_blob(unhappy_text)
```

```python
# AI prompt for Satisfied

happy_prompt = f"""
You are a product analyst.

Below is feedback from users who are generally satisfied with their wireless speaker.
Summarize:
1. Common themes mentioned
2. Minor improvement suggestions
3. Features users appreciate but still want refined

Feedback:
{happy_blob}
"""
```

```python
# AI prompt for Unsatisfied

unhappy_prompt = f"""
You are a product analyst.

Below is feedback from users who are dissatisfied with their wireless speaker.
Summarize:
1. Main pain points and complaints
2. Features causing dissatisfaction
3. Clear improvement recommendations

Feedback:
{unhappy_blob}
"""
```

```python
api_key = userdata.get("GEMINI_API_KEY")
client = genai.Client(api_key=api_key)
```

```python
test = client.models.generate_content(
    model="gemini-2.5-flash",
    contents="Reply with READY"
)
print(test.text)
```

```python
happy_response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=happy_prompt
)

print(happy_response.text)
```

```python
unhappy_response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=unhappy_prompt
)

print(unhappy_response.text)
```

```python
happy_blob = build_text_blob(happy_text)
unhappy_blob = build_text_blob(unhappy_text)

wc = WordCloud(width=900, height=450, background_color="white")

plt.figure(figsize=(12,5))
plt.imshow(wc.generate(happy_blob))
plt.axis("off")
plt.title("Highly Satisfied — Improvement Suggestions")
plt.show()

plt.figure(figsize=(12,5))
plt.imshow(wc.generate(unhappy_blob))
plt.axis("off")
plt.title("Unsatisfied — Improvement Suggestions")
plt.show()
```
